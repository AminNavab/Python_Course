# (sort)


```python

#1)
def bubble_sort(a):
    for i in range(len(a)-1, 0 , -1):
        for j in range(i):
            if a[j] > a[j+1]:
                t = a[j]
                a[j] = a[j+1]
                a[j+1] = t
b = [1,5,4,9,7,6,2]       
bubble_sort(b) 
print(b)     



#2)
def insertion_sort(a):
    for i in range(1,len(a)):
        cv = a[i]
        p = i
        while p>0 and a[p-1]>cv:
            a[p]=a[p-1]
            p = p-1
            a[p] = cv
b = [1,5,4,9,7,6,2] 
insertion_sort(b)
print(b)     
#3)
def selection_sort(a):
    for i in range(len(a)-1,0,-1):
        p=0
        for j in range(1,i+1):
            if a[j]>a[p]:
                p = j 
        t = a[i]
        a[i]=a[p]
        a[p]=t              
b = [1,5,4,9,7,6,2]   
selection_sort(b)
print(b)
```

# matrix

```python
# matrix
# A+B=?
# A-B=?
# A*B=?
# transpose
a = int(input("enter your satre(matrix 1):\n"))
b = int(input("enter your seton(matrix 1):\n"))
A = []
for i in range(0,a):
    A_1 = []
    for j in range(0,b):
        c = int(input("enter number:\n"))
        A_1.append(c)
    A.append(A_1)    
print(A)

a_1 = int(input("enter your satre(matrix 2):\n"))
b_1 = int(input("enter your seton(matrix 2):\n"))
B = []
for i in range(0,a_1):
    B_1=[]
    for j in range(0,b_1):
         c_1 = int(input("enter number:\n"))
         B_1.append(c_1)
    B.append(B_1)     
print(B)

# A+B=?
d = []
for i in range(a):
    d_1 = []
    for j in range(b):
        x = A[i][j]+B[i][j]
        d_1.append(x)
    d.append(d_1)    

print(d)
        
# A-B=?
e = []
for i in range(a):
    e_1 = []
    for j in range(b):
        y = A[i][j]-B[i][j]
        e_1.append(y)
    e.append(e_1)    

print(e)
print("----------------------------------")

# A*B=?
import numpy as np
A = np.array(A)
B = np.array(B)
z = A.dot(B)
print(z)
print("--------------------------------------")

# transpose
print(A.transpose())
print(B.transpose())
```

# union - difference - intersection

```python
# make list
def make_list():
    num = input("enter elemants:\n")
    a = num.split()
    return a
a = make_list()
b = make_list()
print(a)
print(b)

# intersection
def intersection(a,b):
    c = []
    for i in a:
        if i in b:
            c.append(i)
    return c
c = intersection(a,b)
print(c)        

# difference(a-b)
def difference(a,b):
    d = []
    for i in a:
        if i not in b:
            d.append(i)
    return d
d = difference(a,b)
print(d)

# difference(b-a)
def difference(a,b):
    d = []
    for i in b:
        if i not in a:
            d.append(i)
    return d
d = difference(a,b)
print(d)

# union
def union(a,b):
    e = []
    for i in a:
        if i in b:
            e.append(i)
        else:
            e.append(i)
    for j in b:
        if j not in a:
            e.append(j)
    return e
e = union(a,b)
print(e)                
```
